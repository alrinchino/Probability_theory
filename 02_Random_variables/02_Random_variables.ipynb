# Тема 5. Случайные величины



## Понятие случайной величины
Понятие случайной величины – одно из важнейших в теории вероятностей.

Под случайной величиной понимают величину, которая в результате опыта со случайным исходом принимает то или иное числовое значение, причём заранее неизвестно, какое именно.

Случайные величины (кратко СВ) обычно обозначают прописными латинскими буквами
$$
X, \; Y, \; Z\dots ,
$$
иногда используются [«греческие»](https://ru.wikipedia.org/wiki/%D0%93%D1%80%D0%B5%D1%87%D0%B5%D1%81%D0%BA%D0%B8%D0%B9_%D0%B0%D0%BB%D1%84%D0%B0%D0%B2%D0%B8%D1%82) обозначения
$$
\xi - \text{"кси"}, \ \ \theta  - \text{"тета"}, \ \ \zeta  - \text{"дзета"},\dots,
$$
а принимаемые ими значения – малыми буквами
$$ x_1,\, x_2,\,\dots,\, y_1,\,y_2,\,\dots,$$
$$ \xi_1,\, \xi_2,\,\dots,\, \theta_1, \, \theta_2, \, \dots.$$

Используя теоретико-множественную трактовку, можно дать более строгое определение.

Случайная величина $X$ есть действительная числовая функция, определенная на пространстве элементарных исходов $\Omega = \{\omega_1, \;  \omega_2, \; ..., \; \omega_n\} $.

Таким образом, с.в. $X$ каждому элементарному событию $\omega $ ставит в соответствие действительное число $X(\omega ),$ т. е.

$$ X: \ \ \Omega \to \mathbb{D}$$
$$ X \left( \omega  \right) = x,$$
$$ \omega \in \Omega, \ x\in \mathbb{D}, \ \ \mathbb{D} \subset \mathbb{R}, \  \mathbb {R}^n $$

Случайные величины 	называются одномерными, если
$$X: \ \Omega \to \mathbb{D}\subset \mathbb{R},$$
многомерными, если
$$ X: \ \Omega \to \mathbb{D}\subset \mathbb{R}^n, $$
$$ {{\mathbb{R}}^{n}}=\left\{ \left( {{x}_{1}},{{x}_{2}},...,{{x}_{n}} \right)|{{x}_{i}}\in \mathbb{R} \right\} \text{--- арифметическое n-мерное пространство} $$

Примеры одномерных случайных величин.
1. Количество «решек» при 5-кратном подбрасывании монеты:
$$ \Omega =\left\{ OOOOO, \ OOOOP, \ OOOPO, \ ..., \ OOOPP, \ ..., \ PPPPP \right\} $$
$$ X\left( \Omega  \right)=\left\{ 0, \ 1, \ 2, \ 3, \  4, \ 5 \right\} $$

2. Рост случайного прохожего
$$ \Omega =\left\{ \omega_1, \ \omega_2, \ ..., \ \omega_n \right\}$$
$$ X\left( \Omega  \right)=\left\{ 175, \ 167, \ ..., \ 182 \right\} $$

3. Остаток денежной массы на счетах банка по итогам дня

4. Курс доллара

5. Количество бракованных деталей в партии

6. Температура воздуха

7. Выигрыш игрока

8. Координата точки при случайном её выборе на [0; 1]

Примеры многомерных случайных величин
1. Рост, вес, возраст и другие антропометрические данные случайного человека
$$ \Omega =\left\{ \omega_1, \omega_2, \ ..., \ \omega_n \right\} $$

$$ X\left( \Omega  \right)=\left\{ (175, 65, 25, ... ), ( 167, 55, 30, ....), ..., (182, 75, 37, ... ) \right\} $$

2. Координата точки при случайном её выборе на [0; 1]х[0; 1]:
$$ \Omega = \left\{ \omega_1, \ \omega_2, \dots, \omega _n \right\} $$
$$ X\left( \Omega  \right)=\left\{ \left( x,y \right)|0 \leqslant x \leqslant 1, \  0 \leqslant y \leqslant 1 \right\} $$

## Дискретные и непрерывные случайные величины

Если множество возможных значений с.в. $X$ конечно или счётно (это значит, что его элементы могут быть перенумерованы натуральными числами), т. е. дискретно:
$$\mathbb{D}=\left\{ x_1, x_2, \ldots x_n, \ldots | x_i \in \mathbb{R}, \; \mathbb{R}^n \right\},$$
то с.в. $X$ называется дискретной (коротко: ДСВ $X$).

Если же множество значений с.в. $X$ заполняет (непрерывно) конечный или бесконечный промежуток на числовой оси, то такая случайная величина называется непрерывной (коротко: НСВ $X$).
$$\mathbb{D} = \left\{ x \; | \; x_{\min} \leqslant x \leqslant x_{\max }, \; x\in \mathbb{R}, \; \mathbb{R}^n \right\}.$$
О непрерывных случайных величинах пойдёт речь в следующем разделе.

Очевидно, что число возможных значений непрерывной случайной величины бесконечно.

## Закон распределения случайной величины

Для того, чтобы получить полное представление о данной случайной величине, недостаточно знать, какие значения она принимает – важно еще знать, насколько часто они принимаются этой величиной («выпадают») в результате испытаний. Для этой цели используют понятие *закона распределения*.

Любое правило (таблица, функция, график), позволяющее находить вероятности произвольных событий $A\in \Omega $ (в частности, вероятности того, что данная с.в. примет конкретное значение или попадёт в заданный интервал), называется **законом распределения случайной величины** (или короче: **распределением**).

Если для с.в. $X$ задан закон распределения, то говорят, что она распределена по этому закону.

**Закон распределения ДСВ $X$** удобно задавать с помощью следующей таблицы, называемой **рядом распределения**.

|      $X$ | $\mathbb{P}$ |
|:--------:|:--------:|
| $x_1$    | $p_1$   |
| $x_2$    | $p_2$   |
| ...      | ...   |
| $x_n$    | $p_n$   |

или

|      $X$ | $x_1$ | $x_2$ | ... |$x_n$ |
|:--------|:--------:|:--------:|:--------:|:--------:|
| $\mathbb{P}$ | $p_1$   |$p_2$   |...|$p_n$   |


При этом возможные значения $x_1, x_2, ... $ в верхней строке этой таблицы располагаются в определённом порядке, а в нижней – соответствующие вероятности:

$$p_i = \mathbb{P}(X = x_i),$$
причём
$$\sum_i p_i =1.$$
Тем самым определяется **функция вероятности** ***PMF - Probability Mass Function***  – дискретная функция ${{p}_{}}\left( x \right)$ дискретного аргумента, возвращающая вероятность того, что ДСВ $X$ примет определённое значение:

$$PMF(X) = \{\mathbb{P}(X = x_k), \; k = 1,2,...,n\}$$

Графически ряд распределения изображают в виде **многоугольника (полигона) распределения**.

**Замечание.** Поскольку, как было уже отмечено, число возможных значений непрерывной случайной величины бесконечно, задать закон распределения такой величины в виде таблицы или перечисления всех значений не представляется возможным.

Для задания распределения непрерывной случайной величины используются две функции.

**Функция распределения (куммулятивная или интегральная) -** ***Cumulative Distribution Function, CDF***:
$$F(x) \overset{def}{\mathop{=}} \mathbb{P} \{X \leqslant x\} \; \forall{x}.$$

И **функция плотности вероятности распределения (дифференциальная) -** ***Probability Density Function, PDF***:
$$f(x) = F'(x)$$
или
$$f(x): \; F(x) = \int\limits_{-∞}^{+∞} f(x) dx.$$  

## Функция распределения
Кумулятивной функцией распределения (просто функцией распределения, интегральной функцией распределения) СВ $X$ называется функция ${{F}_{X}}(x)$ (коротко $F(x)$), которая для любого числа $x\in \mathbb{R}$ равна вероятности события $\{X\leqslant x\}$ (что случайная величина $X$ примет значение, не превышающее число $x$), т. е.
$$
F(x) \overset{def}{\mathop{=}} \mathbb{P} \{X \leqslant x\} \; \forall{x}.
$$

Функция распределения ДСВ имеет вид
$$F(x) = \sum_{x_i < x}p_i$$
где суммирование ведётся по всем индексам $i$, для которых $x_i < x$.

Функция распределения обладает следующими свойствами:

1. $0 \leqslant F(x) \leqslant 1;$

2. $F(x)$ – неубывающая функция, т.е. $F({{x}_{1}})\leqslant F({{x}_{2}})$, если ${{x}_{1}} \leqslant {{x}_{2}};$

3. $F(-\infty )=0;$ $F(+\infty )=1;$

4. $F(x)$ непрерывна cлева в любой точке $x$, т. е. $F(x-0)=F(x), \: x\in \mathbb{R};$

5. $\mathbb{P} \{\alpha \leqslant X<\beta \}=F(\beta )-F(\alpha )$.

В последнем соотношении неравенства могут быть как строгими, так и нестрогими:
$$ \mathbb{P} (\alpha <X<\beta )= \mathbb{P} (\alpha \leqslant X \leqslant \beta ) = F(\beta) - F(\alpha).$$

## Замечание. Функция **random.seed()**

Данная функция используется для генерации одних и тех же случайных чисел при многократном выполнении кода на одной или разных машинах.

Модуль  random  на  самом  деле  генерирует  псевдослучайные  (т. е.  детерминирован­ные) числа на основе внутреннего состояния. Когда требуется получать воспроиз­ водимые  результаты,  внутреннее  состояние  можно  задавать  при  помощи  посева начального значения с помощь **random.seed()**

`random.seed(svalue, version)`

Параметр  `svalue`  является необязательным, и это начальное значение, необходимое для генерации случайного числа. Значение по умолчанию &mdash; `None`, и в таком случае генератор также использует текущее системное время.

Если вы дважды используете одно и то же начальное значение, вы получите один и тот же результат

```python
np.random.seed(10)
print(random.random())

np.random.seed(10)
print(random.random())
```

Другой пример, в котором мы генерируем одно и то же случайное число много раз

```python
for i in range(5):
    # устанавливаем опцию генератора
    np.random.seed(11)

    # генерируем число от 1 до 1000.
    print(np.random.randint(1, 1000))
```

## Моделирование случайных величин  с помощью функций библиотеки `scipy.stats`

По ссылке можно прочитать полную документацию по работе с непрерывными (Continuous), дискретными (Discrete) и многомерными (Multivariate) распределениями. Конкретное распределение с параметрами `params` можно получить как **`X(params)`**. У него доступны следующие методы:

* `X(params).rvs(size=N)` — генерация выборки размера $N$ (<b>R</b>andom <b>V</b>ariate<b>S</b>). Возвращает `numpy.array`;
* `X(params).cdf(x)` — значение функции распределения в точке $x$ (<b>C</b>umulative <b>D</b>istribution <b>F</b>unction);
* `X(params).logcdf(x)` — значение логарифма функции распределения в точке $x$;
* `X(params).ppf(q)` — $q$-квантиль (<b>P</b>ercent <b>P</b>oint <b>F</b>unction);
* `X(params).mean()` — математическое ожидание;
* `X(params).median()` — медиана ($1/2$-квантиль);
* `X(params).var()` — дисперсия (<b>Var</b>iance);
* `X(params).std()` — стандартное отклонение = корень из дисперсии (<b>St</b>andard <b>D</b>eviation).

Кроме того для непрерывных распределений определены функции
* `X(params).pdf(x)` — значение плотности в точке $x$ (<b>P</b>robability <b>D</b>ensity <b>F</b>unction);
* `X(params).logpdf(x)` — значение логарифма плотности в точке $x$.

А для дискретных
* `X(params).pmf(k)` — значение дискретной плотности в точке $k$ (<b>P</b>robability <b>M</b>ass <b>F</b>unction);
* `X(params).logpdf(k)` — значение логарифма дискретной плотности в точке $k$.

Все перечисленные выше методы применимы как к конкретному распределению `X(params)`, так и к самому классу `X`. Во втором случае параметры передаются в сам метод. Например, вызов `X.rvs(size=N, params)` эквивалентен `X(params).rvs(size=N)`.

Параметры могут быть следующими:
* `loc` — параметр сдвига;
* `scale` — параметр масштаба;
* и другие параметры (например, $n$ и $p$ для биномиального; $a$, $b$ для равномерного и т.д.).

Заметим, что rvs генерирует псевдослучайные числа из любого распределения: uniform,norm,poisson,bernoulli,expon,lognorm
norm,t,chi2,f

```python
rv_unif = st.uniform.rvs(size=10)
print(rv_unif)

rv_norm=st.norm.rvs(loc = 5, scale = 1, size =(10,10))
print(rv_norm)

rv_beta=st.beta.rvs(size=10,a=4,b=2)
print(rv_beta)
```

Для этого надо создать класс, который будет наследоваться от класса **rv_discrete** для дискретных случайных величин и от класса **rv_continuous** для непрерывных случайных величин.
<a  target="_blank" href="http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_continuous.html#scipy.stats.rv_continuous">Пример из документации.</a>

**rv_discrete** является базовым классом для создания конкретных классов распределения и экземпляров дискретных случайных величин. Его также можно использовать для построения произвольного распределения, определяемого списком опорных точек и соответствующих вероятностей.

Если дискретная случайная величина может принимать небольшое число значений, то можно не создавать новый класс, как показано выше, а явно указать эти значения и из вероятности.

```python
x = [-2, 1, 2, 3] # список значений
p = [0.08, 0.40, 0.32, 0.2] # список вероятностей
X = st.rv_discrete(name = 'discrete_distribution',
    values=(x, p)  # значения и вероятности
)

F = 0
for i, j in enumerate(x):
  print(f'при x < {i} F(x) = {F}')
  F +=p[i]
```

### Среднее значение (простая средняя), математическое ожидание (Expectation)

**Математическое ожидание** дискретной величины -- сумма всех значений измеренного признака делится на количество измеренных значений:
$$\mathbb{E}(X) = \frac1N \sum_{i=1}^{N}x_i.$$

#### Свойства математического ожидания:
* $ \mathbb{E}(c) = c, \: c =\operatorname{const} \text{(в общем случае $c$ -- неслучайная величина)}$;
* $ \mathbb{E} \left(c \cdot X\right)=c\cdot \mathbb{E} \left( X \right)$;
* $ \mathbb{E} \left( X+Y \right)= \mathbb{E} \left( X \right)+ \mathbb{E} \left( Y \right)$;
* $ \mathbb{E} \left( \sum \nolimits_{i=1}^{n}{X_i} \right)= \sum \nolimits_{i=1}^{n} {\mathbb{E} \left( X \right)}$;
* Если $X$ и $Y$ – независимы (некоррелированы), то
  $\mathbb{E} \left( X\cdot Y \right)= \mathbb{E} \left( X \right)\cdot \mathbb{E} \left( Y \right)$;
  
* $\mathbb{E} \left( \prod \nolimits_{i=1}^{n} {X_i} \right) = \prod \nolimits_{i=1}^{n} \mathbb{E} \left( X_i \right)$;
* $\mathbb{E} \left [ L \left( X \right) \right] = \sum\nolimits_{i=1}^{n}{ L\left( {{x}_{i}} \right){{p}_{i}}}$, где $L$ - линейная функция;
* Если $X$ и $Y$ – зависимы (коррелированы), то $\mathbb{E} \left( X\cdot Y \right)= \mathbb{E} \left( X \right)\cdot \mathbb{E} \left( Y \right)+\operatorname{cov}\left( X,Y \right)$.

Выборочным аналогом математического ожидания является **среднее значение**:
$$\bar{x} = \frac1n \sum_{i=1}^{n}x_i.$$

```python
mean = x.mean()
print(f'Средняя = {mean}') #f-строка для вывода результатов на экран
print(f'Среднее значение {x.mean()}')
print('Средняя = ', mean)  #то же самое
print('Средняя равна', np.mean(x)) #либо так

def mean(x):
    s = 0
    for i in x:
        s += i
    return s/len(x)
print('Средняя равна', mean(x))
```

## Усеченное среднее значение (от англ. 'trimmed mean' или 'truncated mean')

Усеченное среднее значение часто используется в спортивных соревнованиях, когда самые низкие и самые высокие оценки судей отбрасываются при подсчете оценки участника.

### Дисперсия (вариация, Variance)

Это средний квадрат отклонений индивидуальных значений признака от их средней величины

$$\mathbb{D}(X) = \mathbb{E} \left(X- \mathbb{E}(X) \right)^2 = \mathbb{E}(X^2) - \mathbb{E}^2(X).$$

Другие обозначения дисперсии:
$$ \mathbb{Var}, \, \sigma^2. $$

Для дискретных величин:
$$\mathbb{D}(X) = \frac1n \sum_{i=1}^{n} \left( x_{i} - \mathbb{E}(x) \right)^2 = \frac1n \sum_{i=1}^{n} x_{i}^2 - \mathbb{E}^2(x). $$

#### Для выборки:
- обычная (смещённая) дисперсия
$$s^2 = \frac1n \sum_{i=1}^{n} (x_{i} - \bar{x})^2 = \frac1n \sum_{i=1}^{n} x_{i}^2 - \bar{x}^2. $$

- исправленная (несмещённая) дисперсия
$$\hat{s}^2 = \frac1{n-1} \sum_{i=1}^{n} (x_{i} - \bar{x})^2 = \frac1{n-1} \sum_{i=1}^{n} x_{i}^2 - \bar{x}^2. $$

#### Ликбез: Почему именно квадрат, а не модуль или куб?
 Могу предположить, что линейное отклонение более чувствительно выбросам, квадратичное менее, кубическое — ещё менее чувствительно.
 Попробовал посчитать для 3-х выборок: [1,2,3,4,5], [1,2,3,4,50] и [1,2,3,4,500]:

 - Линейное: 2.5, 452.5 и 49502.5
 - Квадратичное: 1.58, 21.27 и 222.49
 - Кубическое: 1.36, 7.68, и 36.71

Модуль не берут потому, что модуль - не гладкая функция. В нуле у модуля имеется "излом" из-за которого у производной происходит разрыв.
А очень многие математические теоремы, которые наверняка потребуются дальше, работают только на гладких функциях.

Вообще, с не гладкими функциями работать не любят. Там все становится сложнее. Поэтому берется квадрат.
[Source](#https://stepik.org/lesson/8076/step/5?discussion=49741&unit=1356)

#### Свойства дисперсии

$$ \mathbb{D}(x+c) = \mathbb{D}(x) $$
$$ \mathbb{D}(x \cdot c) = c^2\mathbb{D}(x) $$


```python
s2 = sum ((x - np.mean(x)) ** 2) / n
print(f'Дисперсия равна {float(s2)}')
print(f'Дисперсия равна {np.var(x)}') # то же самое

#или так
def variance(x):
    mu = mean(x)
    n = len(x)
    n = n-1 if n in range(1, 30) else n
    square_deviation = lambda x : (x - mu) ** 2
    return sum(map(square_deviation, x) ) / n
```

#### Несмещённая (исправленная) дисперсия
$$ \hat{s}^2 = \frac1{n-1}\sum\limits_{i=1}^{n} {\left( x_i - \bar{x} \right)^2} = s^2 \cdot \frac{n}{n-1}.$$
Более устойчивый показатель, обычно используется при $n > 30$

```python
s2_hat = sum ((x - np.mean(x)) ** 2) / (n-1)
print(f'Несмещенная дисперсия равна {s2_hat}')
print(f'Несмещенная дисперсия равна {s2 * n/(n-1)}')
print(f'Несмещенная дисперсия равна {np.var(x, ddof = 1)}')
```

### Стандартное отклонение

$$ \sigma(X) = \sqrt{\mathbb{D}(X)}$$

Выборочное cтандартное отклонение
* обычное (смещённое):
$$ s(x) = \sqrt{s^2(x)}$$
* исправленное (несмещённое):
$$ \hat{s}(x) = \sqrt{\hat{s}^2(x)}$$

```python
print(f'Обычное (смещённое) стандартное отклонение: {np.std(x):.5f}')

print(f'Исправленное (несмещённое) стандартное отклонение: {np.std(x, ddof=1):.5f}')

def stdev(x):
    diffs = 0
    mean = sum(x)/len(x)
    for i in x:
        diffs += (i - mean)**(2)
    return (diffs/(len(x)-1))**(0.5)

print(f'Исправленное (несмещённое) стандартное отклонение: {stdev(x):.5f}')
```

## Асимметрия и эксцесс

**Асимметрия** — это мера асимметрии распределения. Это значение может быть положительным или отрицательным.

* Отрицательная асимметрия указывает на то, что хвост находится в левой части распределения, которая простирается в сторону более отрицательных значений.
* Положительная асимметрия указывает на то, что хвост находится на правой стороне распределения, которая простирается в сторону более положительных значений.
* Нулевое значение указывает на то, что в распределении вообще нет асимметрии, что означает, что распределение совершенно симметрично.

$$Skew = \frac{\mu_3}{\sigma^3},$$
где
$$\mu_3 = \mathbb{E}(X-\mathbb{E}(X))^3.$$

**Эксцесс** — это мера того, является ли распределение тяжелым или легким хвостом по сравнению с нормальным распределением .

* Эксцесс нормального распределения равен 3.
* Если данное распределение имеет эксцесс меньше 3, говорят, что оно является игровым , что означает, что оно имеет тенденцию производить меньше и менее экстремальных выбросов, чем нормальное распределение.
* Если данное распределение имеет эксцесс больше 3, говорят, что оно лептокуртическое , что означает, что оно имеет тенденцию производить больше выбросов, чем нормальное распределение.

$$Kurtosis = \frac{\mu_4}{\sigma^4},$$
где
$$\mu_4 = \mathbb{E}(X-\mathbb{E}(X))^4.$$

**Примечание.** Некоторые формулы (определение Фишера) вычитают 3 из эксцесса, чтобы упростить сравнение с нормальным распределением.

$$Kurtosis = \frac{\mu_4}{\sigma^4} - 3.$$

## Биномиальное распределение
$$X \sim Bin(n, p)$$:
$$X: 0, 1, ..., n $$
$$ P(X = k) = C_n^k \cdot p^k \cdot (1-p)^{n-k} = \frac{n!}{k! (n-k)!} \cdot p^k \cdot (1-p)^{n-k}$$

$X \sim \operatorname{Bin}(n,p)$

$\text{PMF}: \: p(k)=\dbinom nk {{p}^{k}}{{\left( 1-p \right)}^{n-k}}$, где $\dbinom nk =\frac{n!}{k!(n-k)!}$

$\mathbb{E} =np$

$\operatorname{Var}=npq$

$\mathbb{E} \left( {{X}^{2}} \right)={{\mathbb{E} }^{2}}(X)+\operatorname{Var}X={{(np)}^{2}}+npq=np\left( np+q \right)$

$\text{CDF}{{:}_{{}}}F(x)=\sum\limits_{k=0}^{\left\lfloor x \right\rfloor }{\dbinom nk {{p}^{k}}{{\left( 1-p \right)}^{n-k}}}$

**Пример.**
Смоделируем случайную величину $X \sim Bin(n = 10, p = 0.5)$.

```python
import math
n = 10
p = 0.5
X = list(range(n+1)) # значения случайной величины
P = [] # список вероятностей

for k in range(n+1):
  P.append(math.factorial(n)/(math.factorial(k)*math.factorial(n-k)) * (p**k) * (1-p)**(n-k))
print(P)
```

```python
# Выведем на экран в виде таблицы (дата-фрейма):
import pandas as pd
df = pd.DataFrame()
df['X'] = X
df['P'] = P
df
```

##### Распределение Бернулли

```python
# объявим функцию bernoulli() с параметрами p, iter
def bernoulli(p, iter = 1):

  # создадим пустой массив
  result = np.array([])

  # в цикле с количеством итераций iter
  for i in range(iter):

      # если значение np.random.rand() в диапазоне [0, 1) меньше или равно p
      if np.random.rand() <= p:

          # запишем в массив result единицу
          result = np.append(result, 1)

      else:

          # в противном случае, запишем ноль
          result = np.append(result, 0)

  # в результате выполнения функции вернем массив result
  return result
```

```python
# вызовем функцию
res = bernoulli(0.7, 10000)

# выясним количество нулей и единиц в получившемся массиве
_, counts = np.unique(res, return_counts = True)

# рассчитаем относительную частоту значений
counts / len(res)
```

```python
# используем столбчатую диаграмму, в качестве названия столбцов передадим '0' и '1',
# в качестве высоты - относительную частоту значений
plt.bar(['0', '1'], counts / len(res), width = 0.98)
plt.show()
```

## Гипергеометрическое распределение

Рассмотрим следующую задачу. Пусть в партии из $N$ изделий имеется $M$ стандартных ($M<N$). Из партии случайно отбирают $n$ изделий (каждое изделие может быть извлечено с одинаковой вероятностью), причем отобранное изделие перед отбором следующего не возвращается в партию (поэтому формула Бернулли здесь не применима).

Обозначим через $X$ случайную величину – число $m$ стандартных изделий среди $n$ отобранных. Тогда возможными значениями $X$ будут
$$0,\text{ }1,\text{ }2,\ldots {{,}_{{}}}\min \left( M,n \right).$$

Используя классическое определение вероятности, получаем, что вероятность того, что среди $n$ отобранных изделий ровно $m$ стандартных будет равна
$$p_m = \mathbb{P}\left( X=m \right)=\frac{C_{M}^{m}\cdot C_{N-M}^{n-m}}{C_{N}^{n}}.$$

Дискретная случайная величина имеет **гипергеометрическое распределение**, если ее закон распределения имеет следующий вид:

| $X$ | 0 | 1 | 2 | ... | $\min(M, \; n)$ |
|-----|---|---|---|-----|-----------------|
|$\mathbb{P}$ | $p_0$ | $p_1$ | $p_2$ | ... | $p_{\min(M, \; n)}$|

$${{p}_{m}}= \mathbb{P} \left( X=m \right)=\frac{C_{M}^{m}\cdot C_{N-M}^{n-m}}{C_{N}^{n}},$$
$$m=0,\text{ }1,\text{ }2,\ldots {{,}_{{}}}\min (M,n).$$

#  Тема 8. Непрерывные случайные величины и их распределения

Случайная величина $X$ называется непрерывной, если в результате испытания она может принять любое значение из некоторого числового промежутка $\left[ a,b \right]$. Таким образом, множество возможных значений СВ $X$ заполняет (непрерывно) конечный или бесконечный промежуток на числовой оси. Говорят, что НСВ $X$ распределена на $\left[ a,b \right]$:
$$
X\sim \left[ a,b \right].
$$

Функция распределения $F\left( x \right)$ непрерывной случайной величины непрерывна на всей числовой оси (иногда этот факт берут в качестве определения НСВ).

Количество возможных значений НСВ бесконечно. Поэтому, в отличие от дискретных случайных величин, вероятность отдельного значения для непрерывной случайной величины равна нулю:
$$
\mathbb{P} \left\{ X= x \right\}= 0, \; \forall x \in \mathbb{R}.
$$

## Плотность вероятности (плотность распределения, функция плотности, 			probability density function, PDF)

Если существует такая функция $f(x) \geqslant 0$, что для произвольных $\alpha < \beta$
$$
\mathbb{P} \left(\alpha \leqslant X \leqslant \beta \right ) = \int \limits_{\alpha}^{\beta} f(x) dx,
$$
то говорят, что случайная величина $X$ (или её распределение вероятностей) имеет плотность $f(x)$.

Часто плотность распределения определяют как производную функции распределения:
$$
f\left( x \right)={F}'\left( x \right).
$$

График плотности распределения $f\left( x \right)$ называется кривой распределения.

Свойства плотности:
1. $\mathbb{P} \left( \alpha \leqslant X\leqslant \beta \right)= \int \limits_{\alpha}^{\beta} \,f\left( x \right)dx$;
2.  $f\left( x \right)\geqslant 0$;
3.  $\int \limits_{-\infty}^{\infty}\,f\left( x \right)dx=1$;
4.  $F\left( x \right)= \int \limits_{-\infty}^{x}\,f\left( t \right)dt$;
5.  $\underset{x\to \pm \infty }{\mathop{\lim }}\,f(x)=0.$


## Числовые характеристики непрерывных случайных величин

**Математическое ожидание** непрерывной случайной величины $X$ с плотностью вероятности $f(x)$ находится по формуле
$$\mu = \mathbb{E}(X)=\int\limits_{-\infty }^{\infty }{x\cdot f(x)\,dx.}$$

**Дисперсией** непрерывной случайная величины с плотностью $f(x)$ является величина
$$\mathbb{D} ( X ) = \int\limits_{-\infty }^{\infty } ( x- \mu )^2 \cdot f(x) \, dx $$
или
$$\mathbb{D} ( X ) = \int\limits_{-\infty }^{\infty }x^2 \cdot f( x ) \, dx - \mu^2.$$

**Стандартное отклонение**
$$ \sigma(X) = \sqrt{\mathbb{D}(X)}$$

**Мода** НСВ $X$ с плотностью $f(x)$ есть то ее значение $Mode (X)$, при котором функция $f(x)$ достигает максимума:
$$Mode (X) = \operatorname{argmax} f(x).$$

**Медиана** случайной величины $X$ (обозначение $Me( X )$) — есть такое ее значение $x_p$, для которого одинаково вероятно, окажется ли с. в. $X$ меньше $x_p$ или больше $x_p$, т. е.
$$\mathbb{P}( X< x_p ) = \mathbb{P} (X > x_p) = 0.5.$$

**КваНтили**. Квантилем уровня $\alpha $ СВ $X$ называется число ${{q}_{\alpha }}$, удовлетворяющее уравнению
$$
\mathbb{P} \left\{ X \leqslant {{q}_{\alpha }} \right\}=\alpha.
$$
Таким образом, ${{q}_{\alpha }}$ является решением уравнения
$$
F\left( {{q}_{\alpha }} \right)=\alpha.
$$
В частности, ${{q}_{0,5}}=\operatorname{Me}\left( X \right)$ – \textit{медиана}.


**Квартили** - это три точки(значения признака), которые делят **упорядочное** множество данных на 4 равные части.

$Q_1, \, Q_2, \, Q_3$ - так называемые кваРтили, т.е. 0.25, 0.5, 0.75-квантили.

$IQR = Q_3 - Q_1$ - **интерквартильный размах (inter quartile range)** - один из показателей вариации.

**Начальным моментом порядка $k$** $(k=0,1,2,... )$ случайной величины $X$ называется число $\alpha_k$, определяемое по формуле
$$ \alpha_k = \int\limits_{-\infty }^{\infty } x^k \cdot f ( x )\,dx.$$

**Центральным моментом порядка $k$** $(k=0,1,2,... )$ называется число , определяемое по формуле
$$ \mu_k = \int\limits_{-\infty }^{\infty } ( x-\mu )^k \cdot f ( x )\,dx.$$

**Асимметрия (Skewness)** — это мера асимметрии распределения. Это значение может быть положительным или отрицательным.
$$Skew = \frac{\mu_3}{\sigma^3},$$
где
$$\mu_3 = \mathbb{E}(X-\mathbb{E}(X))^3.$$

**Эксцесс (Kurtosis)** — это мера "островершинности" .
$$Kurt = \frac{\mu_4}{\sigma^4},$$
где
$$\mu_4 = \mathbb{E}(X-\mathbb{E}(X))^4.$$
Некоторые формулы (определение Фишера) вычитают 3 из эксцесса, чтобы упростить сравнение с нормальным распределением.
$$Kurt = \frac{\mu_4}{\sigma^4} - 3.$$

```python
# Рисует график плотности непрерывного распределения

def show_pdf(pdf,  xmin, xmax, ymax, grid_size, distr_name, **kwargs):
    """
    pdf - плотность
    xmin, xmax - границы графика по оси x
    ymax - граница графика по оси y
    grid_size - размер сетки, по которой рисуется график
    distr_name - название распределения
    kwargs - параметры плотности
    """

    grid = np.linspace(xmin, xmax, grid_size) # формирует список значений аргумента
    plt.figure(figsize=(10, 5))               # размер графика
    plt.plot(grid, pdf(grid, **kwargs), lw=5) # рисует график
    plt.grid(ls=':')                          # рисует сетку
    plt.xlabel('Значения X', fontsize=18)       # подпись по оси Ox
    plt.ylabel('Плотность', fontsize=18)      # подпись по оси Oy
    plt.xlim((xmin, xmax))                    # границы аргумента x
    plt.ylim((None, ymax))                    # границы аргумента y
    title = 'Плотность {}'.format(distr_name) # формат подписи графика
    plt.title(title.format(**kwargs), fontsize=20)  # подпись графика
    plt.show()                                # вывести на экран
```

## Непрерывное равномерное распределение

${U}(a, b)$ — равномерное распределение.

$X\sim \text{U}\left( a,b \right)$

$\text{PDF:   }f(x)=\left\{ \begin{matrix}
	\frac{1}{b-a}~, & x\in \left[ a,b \right];  \\
	0, & x\notin \left[ a,b \right].  \\
\end{matrix} \right.$

$\text{CDF:  }F(x)=\frac{x-a}{b-a}$

$\mathbb{E} =\frac{a+b}{2}$

$\operatorname{Var}=\frac{{{\left( b-a \right)}^{2}}}{12}$

$ \mathbb{P} \left( \alpha \leqslant X\leqslant \beta  \right)=\frac{\beta -\alpha }{b-a}$
  
Параметры в **scipy.stats.uniform**:  
* Параметр ${loc = a}$ определяет начало отрезка, на котором случайная величина равномерно распределена.
* Параметр ${scale = b-a}$ определяет длину отрезка, на котором задана случайная величина. Значение плотности распределения на данном отрезке убывает с ростом данного параметра, то есть с ростом длины этого отрезка. Чем меньше длина отрезка, тем больше значение плотности вероятности на отрезке.

## Нормальное распределение и его плотность

Непрерывная случайная величина $X$ распределена по нормальному закону с параметрами $\mathbb{E}X = \mu$ (математическое ожидание) и $\mathbb{SD}{X} = \sigma$ (стандартное отклонение)
$$X \sim \mathcal{N}(\mu, \sigma)$$
если её функция плотности имеет вид:

$$f(x) = \frac{1}{\sqrt{2\pi}\sigma} exp \left(-\frac{(x-\mu)^2}{2\sigma^2} \right) = \frac{1}{\sqrt{2\pi}\sigma} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

Функцию плотности нормального распределения обозначается **npdf - Normal Probability Density Function**

Функция распределения CDF имеет вид:


$$F(x) = \frac{1}{\sigma\sqrt{2\pi}} \int\limits_{-\infty}^x e^{-\frac{(t-\mu)^2}{2\sigma^2}} dt$$.

Если $\mu = 0, \; \sigma = 1$, то распределение называется **стандартным нормальным**.

Функция плотности стандатного нормального распределения имеет вид:
$$\phi(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}.$$

Функция стандратного нормального распределения:
$$\Phi(x) = \frac{1}{\sqrt{2\pi}} \int\limits_{-\infty}^x e^{-\frac{t^2}{2}} dt.$$

$X\sim \text{N}\left( \mu ,\sigma  \right)$

$\text{PDF}:
f \left( x \right) = \frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{(x-\mu )^2}{2 \sigma^2}} \: x\in \mathbb{R}$

$\text{CDF:   }F\left( x \right)= \int \limits_{-\infty}^{x} \,f(t)dt=\frac{1}{\sqrt{2\pi }\sigma } \int \limits_{-\infty}^{x}\,{{e}^{-\frac{{{(t-\mu )}^{2}}}{2\sigma _{{}}^{2}}}}dt$

$\mathbb{E} =\mu $

$\operatorname{Var}={{\sigma }^{2}}$

$\mathbb{P} \left( \alpha \leqslant X\leqslant \beta  \right)=\text{ }\!\!\Phi\!\!\text{ }\left( \frac{\beta -\mu }{\sigma } \right)-\text{ }\!\!\Phi\!\!\text{ }\left( \frac{\alpha -\mu }{\sigma } \right)$

$\mathbb{P} \left( \left| X-\mu  \right|\leqslant \varepsilon  \right)=2\text{ }\!\!\Phi\!\!\text{ }\left( \frac{\varepsilon }{\sigma } \right)$,
где $\Phi \left( z \right)=\frac{1}{\sqrt{2\pi }}\int \limits_{-\infty}^{x}\, {e}^{-\frac{t^2}{2}}dt$ – функция стандартного нормального распределения.

### Пример. Распределение роста взрослых людей.
Построим график плотности нормлаьного распределения для $\mu = 170$ и $\sigma = 10$.

```python
mu = 170 # математическое ожидание
sigma = 10 # стандартное отклонение

def gauss(mu, sigma, x):
  return 1/(math.sqrt(2*math.pi)*sigma) * np.exp(-(x-mu)**2 / (2*sigma**2) )

x = np.linspace(100, 210, 100)
y = gauss(mu, sigma, x)

plt.plot(x, y)
```

### Параметры $\mu, \; \sigma$ и их влияние на графики pdf

Значения параметров определяют положение и форму кривой на графике распределения, каждой комбинации параметров соответствует уникальное распределение.


Для нормального распределения:
* Параметр $loc = a$ отвечает за смещение кривой вдоль $\mathcal{Ox}$, тем самым определяя положение вертикальной оси симметрии плотности распределения. Вероятность того, что значение случайной величины $x$ попадет в отрезок $\mathcal{[m; n]}$, равна площади участка, зажатого кривой плотности, $\mathcal{Ox}$ и вертикальными прямыми ${x = m}$, ${x = n}$. В точке $a$ значение плотности распределения наибольшее, соответственно вероятность того, что значение случайной величины, имеющей нормальное распределение, попадет в окрестность точки $а$ — наибольшая.

  
* параметр ${scale = \sigma}$ отвечает за смещение экстремума вдоль  $\mathcal{Oy}$ и "прижимание" кривой к вертикальной прямой  ${x = a}$, тем самым увеличивая площадь под кривой плотности в окрестности точки $а$. Другими словами, этот параметр отвечает за дисперсию — меру разброса значений случайной величины. При уменьшении параметра $\sigma$ увеличивается вероятность того, что нормально распределенная случайная величина будет равна $a$. Это соответствует мере разброса значений случайной величины относительно её математического ожидания, то есть дисперсии $\sigma^2$.

### Пример. Построим её графики npdf для различных $\mu$ и $\sigma$

```python
def npdf(x, mu, sigma): #npdf - Normal Probability Density Function
    return (1 /(math.sqrt(2**math.pi)*sigma)) * math.exp(-(x - mu)**2 / (2*sigma**2))

#x = [i/10.0 for i in range(-50, 50)]
x = np.linspace(-5, 5, 100)

plt.figure( figsize = (12, 6) )
plt.plot(x, [npdf(i, 0, 1) for i in x], '--')
plt.plot(x, [npdf(i, 2, 1) for i in x], '*')
plt.plot(x, [npdf(i, 2, 2) for i in x], '.')

plt.show()
```

```python
def show_pdf(pdf,  xmin, xmax, ymax, grid_size, distr_name, **kwargs):
    """
    Рисует график плотности непрерывного распределения

    pdf - плотность
    xmin, xmax - границы графика по оси x
    ymax - граница графика по оси y
    grid_size - размер сетки, по которой рисуется график
    distr_name - название распределения
    kwargs - параметры плотности
    """

    grid = np.linspace(xmin, xmax, grid_size)
    plt.figure(figsize = (12, 5))
    plt.plot(grid, pdf(grid, **kwargs), lw=5)
    plt.grid(ls=':')
    plt.xlabel('Значение', fontsize=18)
    plt.ylabel('Плотность', fontsize=18)
    plt.xlim((xmin, xmax))
    plt.ylim((None, ymax))
    title = 'Плотность {}'.format(distr_name)
    plt.title(title.format(**kwargs), fontsize=20)
    plt.show()
```

# Распределение Стьюдента

Распределением Стьюдента (или $t$-распределением) с $n$ степенями свободы называется распределение случайной величины
$${{t}_{n}}=\frac{{{X}_{0}}}{\sqrt{\frac{X_{1}^{2}+X_{2}^{2}+...+X_{n}^{2}}{n}}}=\frac{{{X}_{0}}\sqrt{n}}{\sqrt{X_{1}^{2}+X_{2}^{2}+...+X_{n}^{2}}},$$
где $${{X}_{0}}{{,}_{{}}}{{X}_{1}}{{,}_{{}}}{{X}_{2}}{{,}_{{}}}...,{{X}_{n}}$$ – независимые случайные величины, распределенные по стандартному нормальному закону:
$${{X}_{i}}\sim N\left( 0;1 \right), i=0,1,2,...,n.$$
Обозначение:
$t\sim St(n)$ или $t\sim t(n)$.
Таким образом, формула Стьюдента для средней арифметической
$t=\frac{\overline{X}-\mu }{{{S}_{{\bar{X}}}}}$
– есть частный случай распределения Стьюдента.

– есть частный случай распределения Стьюдента.


1. Математическое ожидание и дисперсия случайной величины ${{t}_{n}}$ равны:
$$M\left( {{t}_{n}} \right)={{0}_{{}}}(n>1){{;}_{{}}}D\left( {{t}_{n}} \right)={{\frac{n}{n-2}}_{{}}}(n>2).$$
2. Плотность распределения Стьюдента равна:
$${{f}_{St}}(t)=\frac{\Gamma \left( \frac{n+1}{2} \right)}{\Gamma \left( \frac{n}{2} \right)\sqrt{\pi n}}{{\left( 1+\frac{{{t}^{2}}}{n} \right)}^{-\frac{n+1}{2}}},$$
где $\Gamma $ – гамма-функция Эйлера:
$$\frac{\Gamma \left( \frac{n+1}{2} \right)}{\Gamma \left( \frac{n}{2} \right)\sqrt{\pi n}}=\frac{(n-1)(n-3)\cdot \cdot \cdot 5\cdot 3}{\pi \sqrt{n}(n-2)(n-4)\cdot \cdot \cdot 4\cdot 2}$$ для чётных $n$,
$$\frac{\Gamma \left( \frac{n+1}{2} \right)}{\Gamma \left( \frac{n}{2} \right)\sqrt{\pi n}}=\frac{(n-1)(n-3)\cdot \cdot \cdot 4\cdot 2}{\pi \sqrt{n}(n-2)(n-4)\cdot \cdot \cdot 5\cdot 3}$$ для нечётных $n$,
Функция плотности вероятности распределения Стьюдента, как и стандартного нормального распределения, является четной функцией.
3. График плотности распределения Стьюдента, как и нормального распределения, является симметричным и колоколообразным, но с более тяжелыми хвостами – из-за этого реализации случайной величины с распределением Стьюдента чаще сильно отличаются от математического ожидания.
4. Функция распределения Стьюдента имеет вид:
$$F(t)=\frac{\Gamma \left( \frac{n+1}{2} \right)}{\Gamma \left( \frac{n}{2} \right)\sqrt{\pi n}}\int\limits_{-\infty }^{t}{{{\left( 1+\frac{{{y}^{2}}}{n} \right)}^{-\frac{n+1}{2}}}dy}.$$
5. При $n\to \infty $ $t$-распределение приближается к нормальному. Практически уже при $k>30$ можно считать $t$-распределение приближенно нормальным.



```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sk
```

```python
mu = 15
sigma = 4
N = 500
X = st.norm(mu, sigma).rvs(N)
Z = (X - mu)/sigma
t = (X - np.mean(X))/(np.std(X))
```

### Замечание. Смещённое и несмещённое стандартное отклонение

Стандартное отклонение можно рассматривать либо как параметр $\sigma$, если речь идет о генеральной совокупности, либо как статистику s, если мы говорим о выборке. Оба значения могут быть вычислены по следующим формулам:

$$ \sigma={\sqrt {{\frac {1}{n}}\sum _{i=1}^{n}\left(x_{i}-{\mu}\right)^{2}}};$$ $$s={\sqrt {{\frac {1}{n - 1}}\sum _{i=1}^{n}\left(x_{i}-{\bar {x}}\right)^{2}}}.$$

Как видите, отличия не очень сильные и заключаются только в том, что при расчете $\sigma$ мы используем мат. ожидание генеральной совокупности - $\mu$ и делим на $n$, а при вычислении sиспользуем среднее значение выборки и делим на $n - 1$. Зачем делить на $n - 1$ вместо $n$? Все дело в том, что при расчете отклонения выборки вместо $\mu$ используется $\bar{x}$ из-за чего значения sполучаются несколько ниже, чем реальное значение $\sigma$ для генеральной совокупности. Поэтому знаменатель $n - 1$ нужен для того, чтобы несколько подправить результат - пододвинуть его чуть поближе к истинному значению.

Для того, чтобы учитывать поправку в методе **std()** пакета **NumPy** есть параметр ddof, который по умолчанию равен 0, но если мы применяем std() к небольшой выборке, то необходимо явно указать, что ddof=1. Влияние данного параметра на расчеты может быть очень значимым. Например, если мы возьмем 10000 выборок по 10 элементов из $N(80, 5^{2})$ и построим распределение стандартных отклонений, то увидим, что при ddof=0 пик распределения находится левее реального значения стандартного отклонения. При ddof=1 пик распределения так же расположен левее реального значения, но все-таки ближе к нему, чем при ddof=0:

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (12, 5))

for i in [0, 1]:
    deviations = np.std(st.norm.rvs(80, 5, (10000, 10)), axis=1, ddof=i)
    sns.histplot(x=deviations ,stat='probability', ax=ax[i])
    ax[i].vlines(5, 0, 0.06, color='r', lw=2)
    ax[i].set_title('ddof = ' + str(i), fontsize = 15)
    ax[i].set_ylim(0, 0.07)
    ax[i].set_xlim(0, 11)
fig.suptitle(r'$s={\sqrt {{\frac {1}{10 - \mathrm{ddof}}}\sum _{i=1}^{10}\left(x_{i}-{\bar {x}}\right)^{2}}}$',
             fontsize = 20, y=1.15);
```

Является ли этот факт важным и можно ли теперь быть уверенными в выводе, который мы сделали на основе Z-статистики? Первое, что приходит в голову - посмотреть, как может быть распределено стандартное отклонение большого количества выборок из генеральной совокупности. Давайте сделаем 5000 тысяч выборок по 10 человек, взятых из распределения N(80, 5) и посмотрим на распределение стандартного отклонения баллов этих выборок:

```python
deviations  = np.std(st.norm.rvs(80, 5, (5000, 10)), axis=1, ddof=1)
sns.histplot(x=deviations ,stat='probability')
```

Это похоже на колокол нормального распределения, но важнее то, что стандартное отклонение выборочных оценок не отклоняется до значения равного 10-и баллам. А это крайне настораживающе. Судите сами, если вероятность того, что отклонение среднего балла наших 10 респондентов от среднего генеральной совокупности составляет всего 2%, то вероятность того, что стандартное отклонение (уж простите за каламбур) может отклониться до значения в 10 баллов стремится к 0. А это означает, что для наших выводов об улучшении качества пива, есть вполне уместное замечание: может быть опрос 10-и респондентов свидетельствует не только об изменении среднего генеральной совокупности, может быть ее стандартное отклонение тоже изменилось.

В конце концов, добавление нового сорта хмеля могло привести к большей специфичности вкуса, следовательно, вызвать большую разнонаправленность в предпочтениях: кому-то этот вкус пива стал нравиться больше, кому-то меньше, но средняя оценка могла остаться прежней. Например, если генеральная совокупность оценок "нового" пива теперь распределена как $N(80, 10^{2})$, то Z-статистика и p-value для имеющихся 10-и оценок будет выглядеть не так обнадеживающе:

```python
z = 10**0.5*(sample.mean()-80)/10
p = 1 - (st.norm.cdf(z) - st.norm.cdf(-z))
print(f'z = {z:.3}')
print(f'p-value = {p:.4}')
```

Оказывается, что для распределения $N(80, 5^{2})$ резульататы оказались значимыми, но как только мы оценили стандартное отклонение генеральной совокупности тем же значением, что и у выборки, т.е. представили, что генеральная совокупность распределена как $N(80, 10^{2})$, то это очень сильно отразилось на значимости имеющихся оценок. Ведь вероятность получить такое же или даже еще более отклоненное среднее теперь равна не 2%, а почти 25%. Хотя, всего-то заменили $\sigma$ на $s$.

В конце концов, не является ли оценка стандартного отклонения генеральной совокупности значением отклонения выборки более разумным решением? Давайте выясним это! Для этого просто сравним то, как могут быть распределены две статистики: (напомню, что параметрами называют характеристики генеральной совокупности, а статистиками - характеристики выборок)

$$Z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}} \; ; \;\;\; T = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}.$$

По сути мы просто придумали новую T-статистику, которая отличается от $Z$-статистики только тем, что в знаменателе стоит не $\sigma$ генеральной совокупности, а $s$ выборки. А теперь сделаем 10000 выборок из $N(80, {5}^2)$, вычислим для каждой выборки $Z-$ и $T-$статистику, а распределение значений изобразим в виде гистограмм:

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (12, 5))

N = 10000
samples = st.norm.rvs(80, 5, (N, 10))
statistics = [lambda x: 10**0.5*(np.mean(x, axis=1) - 80)/5,
              lambda x: 10**0.5*(np.mean(x, axis=1) - 80)/np.std(x, axis=1, ddof=1)]
title = 'ZT'
bins = np.linspace(-6, 6, 80, endpoint=True)

for i in range(2):
    values = statistics[i](samples)
    sns.histplot(x=values ,stat='probability', bins=bins, ax=ax[i])
    p = values[(values > -2)&(values < 2)].size/N
    ax[i].set_title('P(-2 < {} < 2) = {:.3}'.format(title[i], p))
    ax[i].set_xlim(-6, 6)
    ax[i].vlines([-2, 2], 0, 0.06, color='r')
```

Хорошо, значение t-статистики мы вычислили, как вычислить значение p-value? Думаю, вы и до этого уже догадались - нужно проделать все то же самое, что мы делали при вычислении p-value для Z-статистики, но только с использованием t-распределения:

```python
t = st.t(df=9)
fig, ax = plt.subplots()
x = np.linspace(t.ppf(0.001), t.ppf(0.999), 300)
ax.plot(x, t.pdf(x))
ax.hlines(0, x.min(), x.max(), lw=1, color='k')
ax.vlines([-T, T], 0, 0.4, color='g', lw=2)
x_le_T, x_ge_T = x[x<-T], x[x>T]
ax.fill_between(x_le_T, t.pdf(x_le_T), np.zeros(len(x_le_T)), alpha=0.3, color='b')
ax.fill_between(x_ge_T, t.pdf(x_ge_T), np.zeros(len(x_ge_T)), alpha=0.3, color='b')

p = 1 - (t.cdf(T) - t.cdf(-T))
ax.set_title(r'$P(\left | T \right | \geqslant {:.3}) = {:.3}$'.format(T, p))
```

Мы видим, что $p$-value чуть больше 27%, т.е. вероятность того, что результат получен случайно - довольно велика, особенно в соответствии с самым распространенным уровнем значимости $\alpha = 0.05$, который меньше $p$-value более, чем в 5 раз. Что ж, наши результаты не являются значимыми, но они пригодны для демонстрации вычисления доверительного интервала - границы значений математического ожидания генеральной совокупности, в пределах которых оно находится с заданной вероятностью $\alpha$, обычно равной 0.95:

$$\textrm{CI}_{1 - \alpha} = \bar{x} \pm \left ( t_{\frac{\alpha}{2}, \textrm{df}} \right )\left ( \frac{s}{\sqrt{n}} \right )$$

Чтобы вычислить данный интервал в SciPy, достаточно воспользоваться методом interval с заданными параметрами loc (смещение) и scale (масштаб) для нашей выборки:

```python
sample_loc = sample.mean()
sample_scale = sample.std(ddof=1)/10**0.5
ci = st.t.interval(0.95, df=9, loc=sample_loc, scale=sample_scale)
ci
```

Учитывая, что $\bar{x} = 83.7$, то мы можем утверждать, что с вероятностью $\alpha = 0.95$ математическое ожидание генеральной совокупности находится в пределах интервала $[76.5; 90.9]$. Чем шире доверительный интервал, тем больше вероятность того, что мы получим совсем другое выборочное среднее $\bar{x}$, если возьмем другую выборку такого же объема из генеральной совокупности.

## Распределение хи-квадрат

Распределением ${{\chi }^{2}}$ (хи-квадрат) с $n$ степенями свободы называется распределение суммы квадратов $n$ независимых случайных величин $X_{i}^{{}}$, каждая из которых распределена по стандартному нормальному закону:
$$\chi _{n}^{2}=\sum\limits_{i=1}^{n}{X_{i}^{2}},$$
где
$${{X}_{i}}\sim N\left( 0{{;}_{{}}}1 \right), i=1,2,...,n.$$

Число степеней свободы (degrees of freedom, df) у распределения хи-квадрат – это число независимых случайных величин с нормальным стандартным распределением, которые мы суммируем.

Свойства распределения хи-квадрат:
1. Случайная величина, распределенная по закону хи-квадрат, может принимать лишь неотрицательные значения:
$$\chi _{n}^{2}\ge 0$$
2. Функция плотности случайной величины, распределенной по закону хи-квадрат равна:
$${{f}_{\chi _{n}^{2}}}(x)=\frac{1}{{{2}^{\frac{n}{2}}}}\cdot \frac{1}{\Gamma \left( \frac{n}{2} \right)}{{x}^{\frac{n}{2}-1}}{{e}^{\frac{x}{2}}},$$
где $\Gamma $ – гамма-функция Эйлера.
3. Функция распределения $\chi _{n}^{2}$ имеет вид:
$${{F}_{\chi _{n}^{2}}}\left( x \right)=\frac{1}{{{2}^{\frac{n}{2}}}}\cdot \frac{1}{\Gamma \left( \frac{n}{2} \right)}\int\limits_{0}^{x}{{{t}^{\frac{n}{2}-1}}{{e}^{\frac{t}{2}}}dt}.$$
4. Математическое ожидание и дисперсия случайной величины$\chi _{n}^{2}$ равны соответственно:
$$M\left( \chi _{n}^{2} \right)=n,$$
$$D\left( \chi _{n}^{2} \right)=2n.$$


```python
import seaborn as sns

fig, ax = plt.subplots(figsize = (15, 9))
N = 15000
n = 2

x = np.linspace(-5, 5, 100)
ax.plot(x, st.norm.pdf(x, 0, 1), color='r')

samples = st.norm.rvs(0, 1, (N, n))

chi = sum(samples[:, :]**2) #, axis=1)
# t = t[(t>-5) & (t<5)]

sns.histplot(x = t, bins = np.linspace(-5, 5, 100), stat='density', ax=ax)
ax.set_title(r'Форма распределения Стьюдента $t(n)$ при n = ' + str(n), fontsize = 20)
ax.set_xlim(-5, 5)
ax.set_ylim(0, 0.5)
```

## 3. Критерий Колмогорова-Смирнова

Тест Колмогорова-Смирнова используется для проверки того, происходит ли выборка из определенного распределения.

Чтобы выполнить тест Колмогорова-Смирнова в Python, мы можем использовать **scipy.stats.kstest()** для теста с одной выборкой или **scipy.stats.ks_2samp()** для теста с двумя выборками.

Пусть $\alpha$ - уровень значимости (наприример, 0.05, 0.01 и др.)

При $p-value \geqslant \alpha$ нулевую гипотезу $H_0$ о нормальности принимаем.

При $p-value < \alpha $ мы вынуждены отклонить нулевую гипотезу.

### Пример 3.1.
В качестве контрольного примера сгенерируем несколько выборок: нормальную, логнормальную, пуассоновскую, равномерную и проверим их на нормальность.

```python
data1 = np.random.randn(100) # стандартная нормальная выборка объёма 100
data2 = np.random.lognormal(3, 1, 100) # логнормальная выборка
data3 = np.random.poisson(5, 100) # выборка из распределения Пуассона с параметром 5 объёма 100
data4 = np.linspace(-15, 15, 100) #равномерная выборка из интервала (-15, 15) объёма 100
```

```python
ks_stat, p_value = st.kstest(data1, 'norm')
print('Statistics=%.3f, p-value=%.10f' % (ks_stat, p_value))

alpha = 0.05

if p_value >= alpha:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')

ks_stat, p_value = st.kstest(data2,'norm')
print('Statistics=%.3f, p-value=%.10f' % (ks_stat, p_value))
if p_value >= alpha:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')

ks_stat, p_value = st.kstest(data3,'norm')
print('Statistics=%.3f, p-value=%.10f' % (ks_stat, p_value))
if p_value >= alpha:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')

ks_stat, p_value = st.kstest(data4, 'norm')
print('Statistics=%.3f, p-value=%.3f' % (ks_stat, p_value))
if p_value >= alpha:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')
```

## Функциональная и корреляционная зависимости

Исключительный интерес для широкого класса задач представляет обнаружение взаимных связей между двумя и более случайными величинами. Например, существует ли связь между:
*	курением и ожидаемой продолжительностью жизни;
*	между ростом и весом детей;
*	успеваемостью учащихся и результатами выполнения теста IQ;
*	между стажем работы и производительностью труда и т.п.

В инженерных исследованиях такие задачи, обычно, сводятся к установлению связи между некоторым предполагаемым возбуждением и наблюдаемым откликом изучаемой физической системы.

Условно можно вы делить два вида зависимостей между величиной $Y$ и величиной $X$ (или несколькими величинами $X_1,...,X_k$): функциональную и корреляционную.
При функциональной зависимости одна величина зависит только от второй и возникает взаимно-однозначное соответствие: значению одной величины соответствует строго определённое значение второй величины. Такую ситуацию можно описать в виде аналитического выражения (функции):
$$f\left( X \right)=Y$$
или, если представлены выборки,
$$f\left( x_i \right)= y_i, \; i=1,...,n,$$
где
$$x_1,...,x_n\in X; \quad y_1,...,y_n \in Y.$$


Пример 12.1.

1. Функционально связаны, например, общий стаж $Y$ сотрудника и стаж работы на данном предприятии $X$:
$$Y=X+A,$$
где $A$ – стаж работы вне предприятия. Соответственно, для каждого конкретного сотрудника
$$y_i=x_i+a_i.$$
2. Уравнение состояния идеального газа (уравнение Менделеева–Клапейрона) связывает функциональной зависимостью давление, молярный объём и абсолютную температуру идеального газа:
$$ p\cdot V_M = R \cdot T, $$
где $p$  – давление; ${{V}_{M}}$ – молярный объём; $n$ – количество газа; $R$ – универсальная газовая постоянная, $R ≈ 8,314$ Дж/(моль⋅К); $T$ – термодинамическая температура, $К$.
3. Пройденное автомобилем расстояние зависит от скорости и от времени:
$$S=V\cdot t.$$

На практике гораздо чаще встречается корреляционная зависимость, когда распределению значений одной переменной соответствует распределение значений другой переменной, которая подтверждается на одной или нескольких выборках, что позволяет говорить о наличии связи между этими переменными.

Геометрическая иллюстрация данного явления следующая: при корреляционной зависимости результаты наблюдения находятся на некотором приближении к графику функциональной зависимости, но не лежат на нём, а лишь приближаются.

Корреляционная зависимость – возникает в двух случаях:
1. когда одна из величин зависит не только от другой, но и от ряда случайных факторов;
2. обе величины зависят от общих факторов.

Раздел анализа данных, занимающийся изучением корреляционных зависимостей называется корреляционным анализом.

Целью корреляционного анализа является выявление оценки силы связи между случайными величинами (признаками), которые харак-теризует некоторый реальный процесс.


## Парная корреляция. Коэффициент линейной корреляции Пирсона

Рассмотрим, что такое парный коэффициент корреляции. Для этого введем понятие корреляционного момента или ковариации.
По свойству дисперсии двух случайных величин
$$ D(X+Y) = D (X) + D(Y) + 2[\mathbb{E}(XY) - \mathbb{E} (X) \cdot \mathbb{E}(Y)]. $$

Если случайные величины $X$ и $Y$ независимы, то дисперсия их суммы равна сумме их дисперсий
$$D(X+Y) = D(X) + D(Y)$$
и
$$ \mathbb{E}(XY) - \mathbb{E}(X)\cdot \mathbb{E}(Y) = 0.$$
Последнюю величину часто используют как индикатор наличия зависимости между двумя случайными величинами.

Ковариацией (корреляционным моментом) случайных величин $X$ и $Y$ называется число:
$$
\operatorname{cov}\left( X,Y \right)\overset{def} = K_{XY} = \mathbb{E} \left[ \left( X- \mathbb{E} (X \right)\cdot \left( Y- \mathbb{E} (Y \right) \right]= \mathbb{E} \left( X\cdot Y \right)- \mathbb{E} \left( X \right)\cdot \mathbb{E} \left( Y \right),
$$

или, если обозначить $$ \mu_{XY} = \mathbb{E} \left( XY \right), \; \mu_{X} = \mathbb{E} (X), \; \mu_Y = \mathbb{E} ( Y ), $$
то
$$\operatorname{cov} ( X, Y) = \mu_{XY} - \mu_{X} \cdot \mu_Y.$$

Свойства:
1. $\operatorname{cov}\left( X,X \right)=\operatorname{Var}\left( X \right)$;
2. Если $X$ и $Y$ – независимы, то $\operatorname{cov}\left( X,Y \right)=0$;
3. $\operatorname{cov}\left( X,Y \right)=\operatorname{cov}\left( Y,X \right)$;
4. $\operatorname{cov}\left(c \cdot X,Y \right)= c \cdot ~\operatorname{cov}\left( X,Y \right)=\operatorname{cov}\left( X, c\cdot Y \right)$;
5. $\operatorname{cov}\left( X+Y,Z \right)=\operatorname{cov}\left( X,Z \right)+\operatorname{cov}\left( Y,Z \right)$;
6. $\operatorname{cov}\left( X,Y+Z \right)=\operatorname{cov}\left( X,Y \right)+\operatorname{cov}\left( X,Z \right)$.

Если ковариация положительна, то с ростом значений одной случай-ной величины, значения второй имеют тенденцию возрастать, а если знак отрицательный – то убывать.

Однако только по абсолютному значению ковариации нельзя судить о том, насколько сильно величины взаимосвязаны, так как масштаб ковариации зависит от их дисперсий.

Значение ковариации можно нормировать, поделив её на произведе-ние среднеквадратических отклонений (квадратных корней из дис-персий) случайных величин. В этом случае получаем коэффициент корреляции Пирсона.


## Корреляция Пирсона

Корреляция (линейный коэффициент корреляции) или коэффициент корреляции Пирсона двух СВ $X$ и $Y$ определяется как
$$
\rho_{X,Y} \overset{def}{\mathop{=}}\,\frac{\operatorname{cov}\left( X,Y \right)}{\sigma \left( X \right)\cdot \sigma \left( Y \right)}
$$

Зная корреляцию и дисперсии, можно отыскать ковариацию:
$$
\operatorname{cov}\left( X,Y \right) = \rho_{X,Y} \cdot \sigma \left( X \right)\cdot \sigma \left( Y \right)=\rho_{X,Y} \sqrt{\operatorname{Var}X}\cdot \sqrt{\operatorname{Var}Y}.
$$

Свойства:
1. $\rho_{X,Y} = \rho_{Y,X}$;
2. $-1 \leqslant \rho_{X,Y} \leqslant 1$;
3. Коэффициент корреляции равен $ \pm 1$ тогда и только тогда, когда $X$ и $Y$ линейно зависимы (исключая события нулевой вероятности, когда несколько точек «выбиваются» из прямой, отражающей линейную зависимость случайных величин):
$$
\rho_{X,Y} = \pm 1\Leftrightarrow Y=kX+b,k\neq 0, \:
\text{где } k, b \in \mathbb {R}.
$$
Более того в этом случае знаки $\rho_{X,Y}$ и $k$ совпадают:
$\operatorname {sgn} \rho_{X,Y} = \operatorname {sgn} k$.

## Выборочный парный коэффициент корреляции

Выборочный коэффициент корреляции $r$ между двумя выборками количественных данных $x_1,...,x_n$ и $y_1,...,y_n$ определяется по формуле
$$r=\frac{\operatorname{cov}\left( x,y \right)}{{{S}_{x}}\cdot {{S}_{y}}},$$
где
$$S_{x}^{2}=\frac{1}{n}\underset{i=1}{\overset{n}{\mathop \sum }}\,{{\left( {{x}_{i}}-\bar{x} \right)}^{2}}=\frac{1}{n}\underset{i=1}{\overset{n}{\mathop \sum }}\,x_{i}^{2}-{{\left( \frac{1}{n}\underset{i=1}{\overset{n}{\mathop \sum }}\,x_{i}^{{}} \right)}^{2}}=\overline{{{x}^{2}}}-{{\bar{x}}^{2}};$$
$$S_{y}^{2}=\frac{1}{n}\underset{i=1}{\overset{n}{\mathop \sum }}\,{{\left( {{y}_{i}}-\bar{y} \right)}^{2}}=\frac{1}{n}\underset{i=1}{\overset{n}{\mathop \sum }}\,y_{i}^{2}-{{\left( \frac{1}{n}\underset{i=1}{\overset{n}{\mathop \sum }}\,y_{i}^{{}} \right)}^{2}}=\overline{{{y}^{2}}}-{{\bar{y}}^{2}};$$

$$ \operatorname{cov}\left( x,y \right)=\frac{1}{n}\underset{i=1}{\overset{n}{\mathop \sum }}\,\left( {{x}_{i}}-\bar{x} \right)\left( {{y}_{i}}-\bar{y} \right)= $$
$$ =\frac{1}{n}\underset{i=1}{\overset{n}{\mathop \sum }}\,{{x}_{i}}{{y}_{i}}-\frac{1}{n}\underset{i=1}{\overset{n}{\mathop \sum }}\,{{x}_{i}}\cdot \frac{1}{n}\underset{i=1}{\overset{n}{\mathop \sum }}\,{{y}_{i}}=\bar{x}\bar{y}-\bar{x}\cdot \bar{y} .$$

Соответственно
$$ r=\frac{\bar{x}\bar{y}-\bar{x}\cdot \bar{y}}{\left( \overline{{{x}^{2}}}-{{{\bar{x}}}^{2}} \right)\cdot \left( \overline{{{y}^{2}}}-{{{\bar{y}}}^{2}} \right)}=\frac{\frac{1}{n}\sum\limits_{i=1}^{n}{{{x}_{i}}}{{y}_{i}}-\frac{1}{n}\sum\limits_{i=1}^{n}{{{x}_{i}}}\cdot \frac{1}{n}\sum\limits_{i=1}^{n}{{{y}_{i}}}}{\sqrt{\left( \frac{1}{n}\sum\limits_{i=1}^{n}{x_{i}^{2}}-{{{\bar{x}}}^{2}} \right)}\sqrt{\left( \frac{1}{n}\sum\limits_{i=1}^{n}{y_{i}^{2}}-{{{\bar{y}}}^{2}} \right)}}.$$
Если домножить числитель и знаменатель на ${{n}^{2}}$, то
$$r=\frac{n\sum\limits_{i=1}^{n}{{{x}_{i}}}{{y}_{i}}-\sum\limits_{i=1}^{n}{{{x}_{i}}}\cdot \sum\limits_{i=1}^{n}{{{y}_{i}}}}{\sqrt{\left( n\sum\limits_{i=1}^{n}{x_{i}^{2}}-{{{\bar{x}}}^{2}} \right)}\sqrt{\left( n\sum\limits_{i=1}^{n}{y_{i}^{2}}-{{{\bar{y}}}^{2}} \right)}}.$$

### Построение доверительного интервала

```python
# вновь воспользуемся нашими данными о росте
import numpy as np
np.random.seed(42)
height = list(np.round(np.random.normal(180, 10, 1000)))
```

```python
print(height)
```

```python
# теперь нам понадобится модуль stats библиотеки scypy (Scientific Python)
import scipy.stats as st
```

```python
# создадим 90-процентный доверительный интервал (confidence interval, ci)
ci = st.norm.interval(confidence = 0.90, loc = np.mean(height), scale = st.sem(height))

# теперь округлим до трех знаков после запятой и выведем результат
print(np.round(ci, 2))
```

### Проверка гипотезы

```python
# вспомним какое было среднее значение выборки
height_mean = np.mean(height)
print(np.round(height_mean, 2))
```

```python
# теперь будем использовать ttest_1samp, которому передадим нашу выборку и
# предполагаемое истинное среднее 182 см
t_statistic, p_value = st.ttest_1samp(height, 182)

# переменная pvalue как раз и содержит нужную нам вероятность
print(p_value)
```

```python
# теперь напишем небольшую программу, которая скажет отвергнуть нулевую гипотезу или нет
if p_value < 0.05:    # пороговым значением мы выбрали 0,05 или 5%
   print("Есть основания отвергнуть нулевую гипотезу")
else:
  print("Нет оснований отвергнуть нулевую гипотезу")
```
